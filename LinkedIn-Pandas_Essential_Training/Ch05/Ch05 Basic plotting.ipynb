# Basic plotting

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

# Allows to view plot in the notebook itself. 
%matplotlib inline
```

```python
oo = pd.read_csv('../data/olympics.csv',skiprows=4)
oo.head()
```

## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

https://matplotlib.org/examples/

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

### Line plot

```python
fo.Sport.value_counts().plot(kind='line');
```

### Bar plot

```python
# Use the ';' at the end of the line to supress an information line at 
# the top of the chart.
fo.Sport.value_counts().plot(kind='bar');
```

### Horizontal bar plot

```python
fo.Sport.value_counts().plot(kind='barh');
```

### Pie chart

```python
fo.Sport.value_counts().plot(kind='pie');
```

## Plot colors

Color names chart: https://matplotlib.org/examples/color/named_colors.html

```python
fo.Sport.value_counts().plot(color='plum')
```

```python
fo.Sport.value_counts().plot(kind='bar',color='maroon');
```

## figsize()

This allows to expand the figure horizontally so that the X-Axis labels don't overlap.

```python
fo.Sport.value_counts().plot(figsize=(10,3));
```

## Colormaps

```python
fo.Sport.value_counts().plot(kind='pie',colormap='Paired');
```

```python
fo.Sport.value_counts().plot(kind='pie',color='red');
```

# Seaborn basic plotting

Seaborn's example gallery available here: http://seaborn.pydata.org/examples/index.html

```python
import seaborn as sns
```

### How many medals have been won by men and women in the history of the Olympics. How many gold, silver and bronze medals were won for each gender?

```python
sns.countplot(x='Medal',data=oo, hue='Gender');
```

# Challenge

### Plot the number of medals achieved by the Chinese team, so that's both the men and women, in Beijing 2008.
#### Use Matplotlib and Seaborn.

First define a dataframe with Beijing's 2008 data.

```python
dfrBeijing = oo[(oo.Edition == 2008) & (oo.NOC == "CHN")]

dfrBeijing.head(7)
```

```python
dfrBeijing.Medal.value_counts(sort=False).plot(kind='pie',colormap='Paired',figsize=(8,8))
```

```python
sns.barplot(x=["Gold","Bronze","Silver"],
            y=dfrBeijing.Medal.value_counts(),
            palette="Spectral",
            order=['Gold','Silver','Bronze'])
```

### Plot the number of gold, silver, and bronze medals for each gender.



```python
sns.countplot(x='Medal',data=dfrBeijing, hue='Gender', palette='bwr', order=['Gold','Silver','Bronze']);
```
