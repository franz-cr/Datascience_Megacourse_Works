## Question 1

- In which events did **Jesse Owens** win a medal? 


```python
# Import Pandas library
import pandas as pd

# Load data file into a dataframe
dfrOlympics = pd.read_csv("..\\data\\olympics.csv", skiprows=4)

# Display dataframe data to check all is good
dfrOlympics.head(4)
```

### Events where Jesse Owens won a medal

```python
dfrJesseOwens = dfrOlympics[dfrOlympics.Athlete == "OWENS, Jesse"]

dfrJesseOwens.Event.value_counts()
```

```python
dfrJesseOwens[["Event","Medal"]]
```

## Question 2

- Which country has won the most *men's gold medals* in singles badminton over the years? Sort the results alphabetically by the players' names. 

```python
# Boolean indexing
# https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html
# dfrBadmington = dfrOlympics[dfrOlympics["Discipline"] == "Badminton"]
dfrBadmingtonMen = dfrOlympics[(dfrOlympics["Discipline"].str.startswith("Bad")) & 
                               (dfrOlympics.Gender == "Men") &
                               (dfrOlympics.Event == "singles") &
                               (dfrOlympics.Medal == "Gold")]
# Also: dfrOlympics["Discipline"].str.contains("Bad")

dfrBadmingtonMen.sort_values(by="Athlete").head(10)
```

```python
# Count the contries
dfrBadmingtonMen.NOC.value_counts()
```

## Question 3

- Which three countries have won the most medals in recent years, that's from 1984 to 2008? 

```python
dfrRecentYears = dfrOlympics[dfrOlympics.Edition >= 1984]

dfrRecentYears.NOC.value_counts().head(3)
```

## Question 4

- Display the male gold medal winners of the 100-meters track and field sprint event over the years.
    
    List the results starting with the most recent first and show the Olympic city, edition, athlete and the country they represent

```python
dfrMaleGoldWinners = dfrOlympics[(dfrOlympics.Gender == "Men") &
                                 (dfrOlympics.Discipline == "Athletics") &
                                 (dfrOlympics.Medal == "Gold") &
                                 ((dfrOlympics.Event == "100m") | (dfrOlympics.Event == "60m"))
                                  ].sort_values(by="Edition",ascending=False)

dfrMaleGoldWinners[["City","Edition","Athlete","NOC"]]
```

```python
dfrGoldMale100m = dfrOlympics[(dfrOlympics.Medal == "Gold") & 
                              (dfrOlympics.Gender == "Men") & 
                              (dfrOlympics.Event == "100m")]

dfrGoldMale100m.sort_values("Edition", ascending=False)[["City","Edition","Athlete","NOC"]]
```
