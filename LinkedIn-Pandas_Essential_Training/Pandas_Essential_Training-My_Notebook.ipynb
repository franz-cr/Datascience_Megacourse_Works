## Basic Data Analytics

### 'DataFrame.Series.value_counts()'

Method available for series, returns  a tuple object containing counts of the unique values. By **default** results are in descending order so the first row is the one with the most occurrence frequency. Excludes NA values by default. Syntax is:

    Series.value_counts(self, normalize=False, sort=True, ascending=False, bins=None, dropna=True)
    
Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html

```python
# Count of how many medals have been awarded in all Olympic Games
dfrOlympicsDB.Medal.value_counts()
```

```python
# Count of how many medals have been awarded by gender in all Olympic Games
dfrOlympicsDB['Gender'].value_counts(ascending=True)
```

### 'Dataframe.sort_values()':

This function sorts by the value(s) along either axis, since it's working on the entire DataFrame requires the *'by='* parameter to work. The data frame can be sorted by multiple columns when using a list as the *'by='* parameter.

                          by, 
                          axis=0, 
                          ascending=True, 
                          inplace=False, 
                          kind='quicksort',

Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html?highlight=sort_values#pandas.DataFrame.sort_values

### Boolean Indexing (selecting data of a DataFrame)

In boolean indexing a selection of a subset of data is based on the actual values of the data in the DataFrame and not on their row/column labels or integer locations. In boolean indexing, we use a boolean vector to filter the data.

The boolean array/vector returns only **'True'** or **'False'** values which are very useful when passed to the DataFrame selection criteria.

```python
# Boolean array or vector
dfrOlympicsDB.Medal == 'Gold'
```

#### Multiple Conditions in Boolean Indexing

Pandas indexing operators **'&' (AND)**, **'|' (OR)** and **'!' (NOT)** provide easy access to select values from Pandas data structures across a wide range of use cases for multiple boolean conditions. Each condition **must** be enclosed in '(...)' parentheses. 

The array is passed to the DataFrame and only matching rows to the 'True' value for all the boolean expression (having multiple conditions) are returned.

```python
# Boolean array with multiple criteria passed to the DataFrame
dfrOlympicsDB[(dfrOlympicsDB.Medal == 'Gold') & 
              (dfrOlympicsDB.Gender == 'Women')].sample(frac=0.05)
```

**Colormaps for pie charts**

There are 4 types of colormaps:
- *Sequential:* These colormaps are approximately monochromatic colormaps varying smoothly between two color tones---usually from low saturation (e.g. white) to high saturation (e.g. a bright blue). Sequential colormaps are ideal for representing most scientific data since they show a clear progression from low-to-high values.
- *Diverging:* These colormaps have a median value (usually light in color) and vary smoothly to two different color tones at high and low values. Diverging colormaps are ideal when your data has a median value that is significant (e.g.  0, such that positive and negative values are represented by different colors of the colormap).
- *Qualitative:* These colormaps vary rapidly in color. Qualitative colormaps are useful for choosing a set of discrete colors. For example::
        color_list = plt.cm.Set3(np.linspace(0, 1, 12))

    gives a list of RGB colors that are good for plotting a series of lines on a dark background.

Colormaps reference:
https://matplotlib.org/examples/color/colormaps_reference.html

```python
dfrOlympic1980.Sport.value_counts(ascending=False).plot(kind='pie', figsize=(8,8), colormap='tab20');
```

## GroupBy Functions

Groupby is one of the most important functionalities available in Pandas. Groupby does three things:
- It splits a DataFrame into groups based on some criteria, 
- It applies a function to each group independently and 
- It combines the results into a DataFrame. 

Now, many people when they first learn how to use the Groupby function, don't know what to do with the Groupby object. Creating the Groupby object only verifies that you've passed a valid mapping. The **Groupby object isn't a DataFrame but rather a group of DataFrames** in a dict-like structure. So, Groupby splits the DataFrame into groups. Each of these groups is a DataFrame, it applies a function for each group and then finally it combines the results into a DataFrame.

Ref: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

The GroupBy function splits a DataFrame into grouped objects then merges these groups in a new DataFrame with new Series. This is how to GroupBy a Dataframe and the type it is.

```python
grpOlympicsDB = dfrOlympicsDB.groupby('Edition')

print('grpOlympicsDB is of type ', type(grpOlympicsDB))
```

The **grouped by** object is divided by **Edition** and then shows all the rows for that edition, and then for the next edition, and so on.

```python
list(grpOlympicsDB)
```

**Aggregate Function**

Here's a helpful real world tip. Many times you might want to use the aggregation function. The aggregation function, **agg()**, allows multiple statistics to be calculated per group in one calculation. Instructions for aggregation are provided in the form of a python dictionary or a list. 

    DataFrame.groupby(agg([.list.]))
    
    DataFrame.groupby(agg({dict : [...]}))

And the dictionary keys are where you specify which series or columns in your data frame you want to perform the operations and the actual dictionary values specify the function to run. You can also pass custom functions to the list of aggregated calculations and each will be passed the values from the column in your grouped data. Groupby is a very useful Pandas function and it's worth your time making sure you understand how to use it. 

Ref: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html

**Aggregate function using a list**

```python
# Since this DataFrame has mainly 'strings' the 'min', 'max', or 'mean' functions
# do not pull out accurate data.

grpOlympicsDB_02 = dfrOlympicsDB.groupby(['Edition','NOC','Medal'])

grpOlympicsDB_02.agg(['min','max','count']).tail(10)
```

**Aggregate function using a dictionary**

This helps to get more meaninful data. The dictionary allows to perform teh aggreagte functions only on the column specified.

```python
# Since this DataFrame has mainly 'strings' the 'min', 'max', or 'mean' functions
# do not pull out accurate data.

grpOlympicsDB_02.agg({'Medal' : ['min','max','count']}).head(6)
```

Since the *edition series* in our dataset is the only numeric column, let's see if we can get any interesting insights on an athlete such as **Carl Lewis**, who took part in multiple Olympic games. 

So let's look for Carl Lewis and group by athlete. This time we'll **aggregate** on the edition field as this is the only numeric field that we have, so edition, min, max, and count. This is going to provide the first Olympics that he won a medal in, which was in 1984. Remember that we don't know if he took part in any earlier Olympics, this dataset is only for Olympic medal winners. We also have the last Olympics that he took part in which is in 1996, and we see that he's won 10 medals in total.

```python
dfrOlympicsDB.loc[dfrOlympicsDB.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition':['min','max','count']})
```
