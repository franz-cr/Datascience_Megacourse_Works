## Part 1: Bar Charts

Importing library and opening data file.

```python
# Pyplot Function overview
# https://matplotlib.org/3.1.1/api/pyplot_summary.html

import matplotlib.pyplot as plt
import pickle

with open('fruit-sales.pickle', 'rb') as f:
    data = pickle.load(f)
    
print(data)
```

```python
#Split the list of tuples into 2 lists
fruit, num_sold = zip(*data)

print (fruit)
print(num_sold)
```

```python
# Define bar-chart's X-axis
bar_coords = range(len(fruit))

# Define the bar-chart
plt.bar(bar_coords, num_sold)

# Display bar-chart
plt.show()
```

### Chart Annotations

matplotlib.pyplot API Reference: https://matplotlib.org/3.1.1/api/pyplot_summary.html

```python
# matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
plt.bar(bar_coords, num_sold, width=0.9)
plt.title('Total Number of Fruit Sold Today')
plt.ylabel('Quantity')
plt.xlabel('Fruit')
plt.fill(False)
#Replace the coordinates on X-Axis with the actual fruit names
# xticks command receives 2 arrays
plt.xticks(bar_coords, fruit)
plt.show()
```

## Part 3.1: Line Charts

Import, upload and split the data file.

Este archivo de ranking de lenguajes de programación contiene datos en varios niveles:
1) Nivel 1: Lenguaje: "Java"
2) Año: 2018, 2013, 2008, 2003, 1998
3) Ranking para ese año: 1, 2, 1, 1, 16

```python
with open('prog-langs-popularity.pickle', 'rb') as f:
    data = pickle.load(f)
    
print(data)
print(' ')

#Split the list of tuples into 2 lists
languages, rankings = zip(*data)
print('Languages: ', languages)
print(' ')

print('Rankings: ', rankings)
print('Ranking es aun una lista de listas')
print(' ')

java_years, java_ranks = zip(*rankings[0])
print('Java Years: ', java_years)
print('Java Rankings: ', java_ranks)
```

```python
plt.plot(java_years, java_ranks)
plt.title('Java Ranking vrs year')
plt.xticks(java_years)
plt.xlabel('Year')
plt.ylabel('Ranking')

plt.show()
```

## Part 3.2: Multiline Charts

```python
for i in range(len(languages)):
    years, ranks = zip(*rankings[i])
    # matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
    plt.plot(years, ranks)

# Format chart
plt.legend(languages)
plt.title('Ranking of programming languages by year')
plt.xlabel('Year')
plt.ylabel('Ranking')
    
plt.show()
```

```python
# Load the 'iris' dictionary into the 'iris' variable
with open('iris.pickle', 'rb') as f:
    iris = pickle.load(f)
  
print('Iris Dictionary Content: \n', iris,'\n')
```

### Extracting data from  dictionary

Extract the data from the dictionary (\'iris\') only for the \'data\' array and assigning each column to a variable in this order:
\- Sepal length in cm 
\- Sepal width in cm 
\- Petal length in cm 
\- Petal width in cm

```python
# Extracting the sepal length, located in the first column of the 'data' array.
# Using '[:,0]' specifies that all 150 sepal lenghts (':') from first column (index 0) need to be extracted
# Same extract logic used for the sepal width, petal length and width.
sepal_length = iris['data'][:,0]
sepal_width = iris['data'][:,1]
petal_length = iris['data'][:,2]
petal_width = iris['data'][:,3]
# Extracting the target subgroup of data, from the 'iris' dictionary, to a variable
classes = iris['target']
class_names = iris['target_names']

print('Sepal Length: ', sepal_length, '\n')
print('Sepal Width: ', sepal_width, '\n')
print('Petal Length: ', petal_length, '\n')
print('Petal Width: ', petal_width, '\n')
print('Target: ', classes, '\n')
print('Target names: ', class_names, '\n')
```

### Displaying the scatter plot

```python
# matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, 
#    alpha=None, linewidths=None, verts=None, edgecolors=None, *, plotnonfinite=False, data=None, **kwargs)
# Setting up the scatter plot attributes
plt.scatter(sepal_length, sepal_width)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Sepal length and width from Iris dictionary')

plt.show()
```

### Improving the scatter plot visualization

Adding colors to the plot, a color for each class, 
- 'target_names': array(['setosa', 'versicolor', 'virginica'],

```python
# Setting up the scatter plot attributes
plt.scatter(sepal_length, sepal_width, c = classes)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Sepal length and width from Iris dictionary')
plt.legend(class_names)


plt.show()
```

## Part 5: Multiple Plots in one figure


```python
# Declaring the multiple plot figure
fig, axes = plt.subplots(2, 2)

# Defining the first multi-plot figure with the Iris scatter plot
axes[0, 0].scatter(sepal_length, sepal_width, c = classes)
axes[0, 0].set_xlabel('Sepal length')
axes[0, 0].set_ylabel('Sepal width')
axes[0, 0].set_title('Sepals length and width')

# Defining the second multi-plot figure with the Iris scatter plot
axes[0, 1].scatter(petal_length, petal_width, c = classes)
axes[0, 1].set_xlabel('Petal length')
axes[0, 1].set_ylabel('Petal width')
axes[0, 1].set_title('Petals length and width')

# Defining the third multi-plot figure with the bar chart plot
axes[1, 0].barh(dev_types, years_exp)
axes[1, 0].set_xlabel('Years')
axes[1, 0].set_ylabel('Developer')
axes[1, 0].set_title('Developer vrs years')

# Defining the fourth multi-plot figure with the line chart plot
axes[1, 1].plot(java_years, java_ranks)
axes[1, 1].set_xlabel('Java')
axes[1, 1].set_ylabel('Popularity')
axes[1, 1].set_title('Languages ranking')
```

### Example about 'subplots' provided by Mathlib Org 

```python
import numpy as np

# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Creates four polar axes, and accesses them through the returned array
fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axes[0, 0].plot(x, y)
axes[1, 1].scatter(x, y)

# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Creates figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax=plt.subplots(num=10, clear=True)
```

```python
# Coding languages share both X and Y axes with all subplots
fig, axes01 = plt.subplots(2, 2, sharex='all', sharey='all')

# Extract the language rankings
# java_years, java_ranks = zip(*rankings[0])
c_years, c_ranks = zip(*rankings[1])
cpp_years, cpp_ranks = zip(*rankings[2])
python_years, python_ranks = zip(*rankings[3])

# Set line charts up
fig.suptitle('Languages Ranking',fontsize=18, fontweight='demibold')
# weight : {a numeric value in range 0-1000, 'ultralight', 'light', 'normal', 'regular', 
# 'book', 'medium', 'roman', 'semibold', 'demibold', 'demi', 'bold', 'heavy', 'extra bold', 'black'}

axes01[0, 0].plot(c_years, c_ranks)
axes01[0, 0].set_xlabel('C',labelpad=5)
axes01[0, 0].set_ylabel('Popularity')
axes01[0, 0].set_xmargin(50)
axes01[0, 0].set_ymargin(50)

axes01[0, 1].plot(cpp_years, cpp_ranks)
axes01[0, 1].set_xlabel('C++',labelpad=5)

axes01[1, 0].plot(python_years, python_ranks)
axes01[1, 0].set_xlabel('Python',labelpad=5)
axes01[1, 0].set_ylabel('Popularity')

axes01[1, 1].plot(java_years, java_ranks)
axes01[1, 1].set_xlabel('Java',labelpad=5)

```
