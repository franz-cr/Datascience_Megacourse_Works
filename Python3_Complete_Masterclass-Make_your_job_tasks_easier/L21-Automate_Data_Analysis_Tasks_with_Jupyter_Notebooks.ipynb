<h3>Creating a data frame</h3>

<p>A data frame is the object holding the data. To create a data frame it is necessary to pass a list of lists as an argument</p>


#The result, where 0, 1, 2 are the columns and 0, 1, 2, 3 are the indexes

```python
# Import the Pandas library
import pandas as pds

# Define a data structure
dtaMyDataStructure = pds.DataFrame([
    ['Andy', 46, 'Engineer'], 
    ['Jane', 33, 'Nurse'], 
    ['Robert', 21, 'Student'], 
    ['Maria', 30, 'Student']],
    columns = ['Name', 'Age', 'Occupation'], 
    index = ['ID1', 'ID2', 'ID3', 'ID4'],)

# Confirm data structure type
print("Type of 'dtaMyDataStructure': ", type(dtaMyDataStructure), "\n")

# Display the data structure
print("Display 'dtaMyDataStructure': \n\n", dtaMyDataStructure)
```

<h3>Displaying data from the data structure</h3>

<p>Get the series from the table (data structure).</p>
<a href="https://pandas.pydata.org/pandas-docs/stable/">Pandas library documentation</a>

```python
# Display the columns of the table alone
print("Column data in 'Name':\n----------------------\n", 
      dtaMyDataStructure.Name, 
      "\n\nColumn Type: ", 
      type(dtaMyDataStructure.Name), "\n")

print("Column data in 'Age':\n---------------------\n", 
      dtaMyDataStructure.Age, 
      "\n\nColumn Type: ", 
      type(dtaMyDataStructure.Age), "\n")

print("Column data in 'Occupation':\n----------------------------\n", 
      dtaMyDataStructure.Occupation, 
      "\n\nColumn Type: ", 
      type(dtaMyDataStructure.Occupation), "\n")
```

<h3>Filtrar registros de una tabla</h3>
<p>Se pueden extraer tanto un solo registro como varios. https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/03_subset_data.html#how-do-i-filter-specific-rows-from-a-dataframe</p>

```python
# Operadores lógicos accesibles: | -> OR, & -> AND, ~ -> NOT, == -> EQUAL
dfrSales = dfrJSON[dfrJSON["Department"] == "Sales"]

dfrSales
```

```python
dfrNotSalesOrMkt = dfrJSON[~((dfrJSON["Department"] == "Sales") | (dfrJSON["Department"] == "Marketing"))]

dfrNotSalesOrMkt
```

```python
dfrSalaryDepartment = dfrJSON[((dfrJSON["Department"] == "IT") & (dfrJSON["Salary"] < 55000))]

dfrSalaryDepartment
```
