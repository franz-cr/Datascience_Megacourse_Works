### Aggregate Statistics (GroupBy)

This line brings all countries and displays the average per country of all column values

```python
df1 = df

df1.groupby(['cntry']).mean()
```

```python
df1.groupby(['cntry']).sum()
```

```python
df1.groupby(['cntry','year']).count()
```
